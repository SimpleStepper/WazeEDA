# Functions to make Graphs

```python
# Make a function that creates a Boxplot based on the column from the dataset
def plot_box(df,column):

    plt.figure(figsize=(5,1))
    plt.title(f'{column.capitalize()} Box Plot')
    sns.boxplot(x=df[column]);
```

```python
# Make a function that creates a histogram based on the column from the dataset
def plot_histo(df, column, bin_width=25):

# Graph Parameters
    plt.figure(figsize=(5,3))
    sns.histplot(Waze_Data[column], binwidth=bin_width)
    plt.title(f'Number of {column.capitalize()}')
    plt.xlabel(column.capitalize())
    plt.ylabel('Count')

# Calculate and plots Meand and Median values onto graph
    mean_value = df[column].mean()
    plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean:{mean_value:.2f}')
    median_value = df[column].median()
    plt.axvline(median_value, color='orange', linestyle='-', linewidth=2, label=f'Median:{median_value:.2f}')

# display graph with legend

    plt.legend()
    plt.show()
```

# Handling Outliers

```python
# write a function that calculates the 95th percentile of a given column, then imputes values > the 95th percentile with the value at the 95th percentile.
def outlier_cap(df,column,percentile=95):
    threshold = df[column].quantile(percentile / 100) #computes the percentile entered in the function
    df[column] = df[column].clip(upper=threshold) #caps the value according the the threshold percentile using .clip
    print(f'capped values of {column} at the {percentile}th percentile: {threshold:.2f}')
```

```python
columns_to_cap = ['sessions','drives','total_sessions','driven_km_drives','duration_minutes_drives'] # defines what columns to change 

for column in columns_to_cap:
    outlier_cap(Waze_Data, column, percentile=95)
```
